```python
# ================================================================
# STEP 3: Load PDF Document
# ================================================================

import os
from pathlib import Path

# Option 1: Specify PDF file path directly
pdf_path = "/kaggle/input/pdf-to-test-rag/Convolutional Neural Networks (CNNs) in Deep Learning.pdf"  # Change this to your PDF file path

# Option 2: Use file upload (if using Jupyter/Colab)
# from google.colab import files  # Uncomment for Colab
# uploaded = files.upload()
# pdf_path = list(uploaded.keys())[0]  # Get first uploaded file

# Check if file exists
if os.path.exists(pdf_path):
    print(f"âœ“ PDF file found: {pdf_path}")
    print(f"File size: {os.path.getsize(pdf_path) / 1024:.2f} KB")
else:
    print(f"âš ï¸  File not found: {pdf_path}")
    print("Please update 'pdf_path' variable with your PDF file path")
    print("\nExample:")
    print("  pdf_path = 'documents/my_paper.pdf'")
    print("  or")
    print("  pdf_path = '/content/assignment.pdf'  # For Colab")
```

```python
# ================================================================
# STEP 4: Document Loading from PDF
# ================================================================

# Load the PDF document
try:
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()
    
    print(f"âœ“ Loaded {len(documents)} page(s) from PDF")
    print(f"Total characters: {sum(len(doc.page_content) for doc in documents):,}")
    print(f"\nFirst 200 characters from page 1:")
    print(documents[0].page_content[:200] + "...")
    
    # Show page count
    print(f"\nğŸ“„ Document Summary:")
    print(f"  - Total pages: {len(documents)}")
    print(f"  - Average chars per page: {sum(len(doc.page_content) for doc in documents) // len(documents):,}")
    
except Exception as e:
    print(f"âŒ Error loading PDF: {e}")
    print("\nMake sure:")
    print("1. The PDF file path is correct")
    print("2. The file is not corrupted")
    print("3. You have read permissions for the file")
```

```python
# ================================================================
# STEP 5: Text Splitting (Chunking)
# ================================================================

# Split documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,        # Size of each chunk in characters
    chunk_overlap=50,      # Overlap between chunks to maintain context
    length_function=len,
    separators=["\n\n", "\n", " ", ""]
)

chunks = text_splitter.split_documents(documents)

print(f"âœ“ Split into {len(chunks)} chunks")
print(f"\nExample chunk:")
print(f"Chunk 0: {chunks[0].page_content}")
print(f"\nChunk size: {len(chunks[0].page_content)} characters")
```

```python
# ================================================================
# STEP 6: Create Embeddings and Vector Store (FREE Local)
# ================================================================

# AGGRESSIVE protobuf fix
import os
import sys

# Set environment variables BEFORE importing anything
os.environ['PROTOCOL_BUFFERS_PYTHON_IMPLEMENTATION'] = 'python'
os.environ['PROTOBUF_PYTHON_IMPLEMENTATION'] = 'python'

# Try to reload protobuf if needed
try:
    import google.protobuf
    google.protobuf.__version__
except:
    pass

print("Loading FREE local embedding model...")
print("(This will download ~400MB on first run, then cached locally)")

# Try HuggingFace embeddings with error handling
try:
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2",
        model_kwargs={'device': 'cpu'},
        encode_kwargs={'normalize_embeddings': True}
    )
    # Test the embeddings work
    test_embedding = embeddings.embed_query("test")
    print("âœ“ Embedding model loaded and tested!")
    
except Exception as e:
    print(f"âš  Error with HuggingFaceEmbeddings: {e}")
    print("\nTrying alternative: Using OpenAI-compatible embeddings...")
    
    # Alternative: Use a different embedding approach
    try:
        from langchain_community.embeddings import OpenAIEmbeddings
        # This won't work without API key, but shows the structure
        print("âš  OpenAI embeddings require API key")
        raise Exception("Need alternative")
    except:
        pass
    
    # Last resort: Use simple TF-IDF or CountVectorizer approach
    print("âš  Using simple embeddings as fallback...")
    from sklearn.feature_extraction.text import TfidfVectorizer
    from langchain.embeddings.base import Embeddings
    from typing import List
    
    class SimpleTFIDFEmbeddings(Embeddings):
        def __init__(self, documents):
            self.vectorizer = TfidfVectorizer(max_features=384)
            texts = [doc.page_content for doc in documents]
            self.vectorizer.fit(texts)
        
        def embed_documents(self, texts: List[str]) -> List[List[float]]:
            vectors = self.vectorizer.transform(texts).toarray()
            return vectors.tolist()
        
        def embed_query(self, text: str) -> List[float]:
            vector = self.vectorizer.transform([text]).toarray()[0]
            return vector.tolist()
    
    embeddings = SimpleTFIDFEmbeddings(chunks)
    print("âœ“ Using TF-IDF embeddings (simpler, but works!)")

# Create vector store with Chroma
print("\nCreating vector store...")
try:
    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings,
        collection_name="pdf_knowledge_base",
        persist_directory="./chroma_db"
    )
    print(f"âœ“ Vector store created with {len(chunks)} chunks")
except Exception as e:
    print(f"âš  Error creating Chroma vectorstore: {e}")
    print("Trying without persistence...")
    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings,
        collection_name="pdf_knowledge_base"
    )
    print(f"âœ“ Vector store created (without persistence)")

# Test similarity search
test_query = "What is the main topic of this document?"
try:
    similar_docs = vectorstore.similarity_search(test_query, k=3)
    print(f"\nTest Query: '{test_query}'")
    print(f"Found {len(similar_docs)} similar documents")
    print(f"\nMost similar chunk:")
    print(similar_docs[0].page_content[:200] + "...")
except Exception as e:
    print(f"âš  Test search failed: {e}")
    print("Vector store created but test search failed - continuing anyway...")
```

```python
# ================================================================
# STEP 7: Initialize FREE Groq LLM (Very Fast!)
# ================================================================

from langchain_groq import ChatGroq

print("Initializing FREE Groq LLM...")
print("(Groq is VERY fast - often 10x faster than other APIs!)")

try:
    # Groq - FREE, FAST, and reliable!
    llm = ChatGroq(
        model_name="llama-3.1-8b-instant",  # Very fast model
        temperature=0,  # Deterministic responses
        max_tokens=512,
        groq_api_key=os.getenv("GROQ_API_KEY"),  # Get from environment
    )
    print("âœ“ Groq LLM initialized!")
    print("  Model: llama-3.1-8b-instant")
    print("  Temperature: 0 (deterministic)")
    print("  Max tokens: 512")
    print("  âš¡ Very fast responses!")
    
except Exception as e:
    print(f"âŒ Error: {e}")
    print("\nğŸ“‹ To fix:")
    print("1. Get FREE API key from: https://console.groq.com/keys")
    print("2. Sign up (it's free!)")
    print("3. Copy your API key")
    print("4. Set it in Cell 2: GROQ_API_KEY = 'your_actual_key'")
    raise

print("\nâœ“ FREE Groq LLM Setup Complete!")
print("  âš¡ Very fast - works great on Kaggle!")
```

```python
# ================================================================
# STEP 8: Create RAG Prompt Template
# ================================================================

# Create a prompt template for RAG
template = """You are a helpful AI assistant. Use the following context to answer the question.
If you cannot answer the question based on the context, say "I don't have enough information in the provided context to answer this question."

Context:
{context}

Question: {question}

Answer:"""

prompt = ChatPromptTemplate.from_template(template)

print("âœ“ Prompt template created!")
```

```python
# ================================================================
# STEP 9: Build RAG Chain
# ================================================================

# Create retriever
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}  # Retrieve top 3 most similar chunks
)

# Format documents helper function
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Build the RAG chain
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

print("âœ“ RAG chain built successfully!")
```

```python
# ================================================================
# STEP 10: Test the RAG System
# ================================================================

import json
import csv
from datetime import datetime

print("\n" + "="*60)
print("RAG SYSTEM TESTING")
print("="*60)

# Test questions
test_questions = [
    "What are the main components of a CNN?",
    "Which CNN architecture introduced skip connections?",
    "What is the purpose of pooling layers?",
    "What activation function is commonly used in CNNs?",
    "How many parameters does AlexNet have?",
    "What is the difference between max pooling and average pooling?",
    "How does batch normalization improve CNN training?",
    "What are the advantages of using convolutional layers?",
    "Explain the concept of receptive fields in CNNs.",
    "What is data augmentation and why is it important for CNNs?",
    "How does dropout prevent overfitting in neural networks?",
    "What is transfer learning and when should it be used?",
    "Describe the architecture of VGG networks.",
    "What are residual connections and why are they important?",
    "How do 1x1 convolutions reduce computational complexity?"
]

# Collect all answers
all_answers = []

for i, question in enumerate(test_questions, 1):
    print(f"\n{'â”€'*60}")
    print(f"Question {i}: {question}")
    print(f"{'â”€'*60}")
    
    # Retrieve relevant documents
    retrieved_docs = retriever.invoke(question)
    print(f"\nğŸ“š Retrieved {len(retrieved_docs)} relevant chunks:")
    for j, doc in enumerate(retrieved_docs, 1):
        print(f"\nChunk {j}:")
        print(doc.page_content[:200] + "...")
    
    # Generate answer
    print(f"\nğŸ’¡ Answer:")
    answer = rag_chain.invoke(question)
    print(answer)
    print()
    
    # Store answer
    all_answers.append({
        "question_number": i,
        "question": question,
        "answer": answer
    })

# Export to JSON file
json_output_path = "rag_answers.json"
with open(json_output_path, 'w', encoding='utf-8') as f:
    json.dump(all_answers, f, indent=2, ensure_ascii=False)

print(f"\nâœ“ Exported to JSON: {json_output_path}")

# Export to Text file with nice formatting
txt_output_path = "rag_answers.txt"
with open(txt_output_path, 'w', encoding='utf-8') as f:
    f.write("="*70 + "\n")
    f.write("RAG SYSTEM - QUESTION AND ANSWER RESULTS\n")
    f.write("="*70 + "\n")
    f.write(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write("="*70 + "\n\n")
    
    for item in all_answers:
        f.write(f"Question {item['question_number']}: {item['question']}\n")
        f.write("-"*70 + "\n")
        f.write(f"Answer:\n{item['answer']}\n")
        f.write("\n" + "="*70 + "\n\n")

print(f"âœ“ Exported to Text: {txt_output_path}")

# Export to CSV file
csv_output_path = "rag_answers.csv"
with open(csv_output_path, 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['question_number', 'question', 'answer'])
    writer.writeheader()
    writer.writerows(all_answers)

print(f"âœ“ Exported to CSV: {csv_output_path}")

print(f"\nğŸ‰ All answers exported successfully!")
print(f"   Files created:")
print(f"   1. {json_output_path} (JSON format)")
print(f"   2. {txt_output_path} (Text format)")
print(f"   3. {csv_output_path} (CSV format)")
```

```python
# ================================================================
# STEP 11: Interactive Q&A
# ================================================================

def ask_question(question: str):
    """Ask a question and get an answer with sources"""
    print(f"\n{'='*60}")
    print(f"Question: {question}")
    print(f"{'='*60}")
    
    # Retrieve relevant documents
    retrieved_docs = retriever.invoke(question)
    print(f"\nğŸ“š Retrieved Chunks:")
    for i, doc in enumerate(retrieved_docs, 1):
        print(f"\n[Chunk {i}]")
        print(doc.page_content)
    
    # Generate answer
    print(f"\n{'â”€'*60}")
    print(f"ğŸ’¡ Answer:")
    print(f"{'â”€'*60}")
    answer = rag_chain.invoke(question)
    print(answer)
    
    return answer, retrieved_docs

# Example usage
question = "What are the differences between AlexNet and ResNet?"
answer, sources = ask_question(question)
```

```python
# ================================================================
# STEP 12: Evaluation Metrics
# ================================================================

print("\n" + "="*60)
print("RAG SYSTEM EVALUATION")
print("="*60)

def evaluate_retrieval(question: str, relevant_chunks: List[str]):
    """Evaluate retrieval quality"""
    retrieved_docs = retriever.invoke(question)
    retrieved_texts = [doc.page_content for doc in retrieved_docs]
    
    # Calculate precision (how many retrieved docs are relevant)
    relevant_retrieved = sum(1 for text in retrieved_texts if any(rel in text for rel in relevant_chunks))
    precision = relevant_retrieved / len(retrieved_texts) if retrieved_texts else 0
    
    # Calculate recall (how many relevant docs were retrieved)
    recall = relevant_retrieved / len(relevant_chunks) if relevant_chunks else 0
    
    # F1 score
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return {
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "retrieved_count": len(retrieved_texts),
        "relevant_retrieved": relevant_retrieved
    }

# Test evaluation
test_question = "What is ResNet?"
relevant_keywords = ["ResNet", "skip connections", "residual"]

metrics = evaluate_retrieval(test_question, relevant_keywords)
print(f"\nRetrieval Metrics for: '{test_question}'")
print(f"Precision: {metrics['precision']:.2f}")
print(f"Recall: {metrics['recall']:.2f}")
print(f"F1 Score: {metrics['f1_score']:.2f}")
print(f"Retrieved: {metrics['retrieved_count']} chunks")
print(f"Relevant Retrieved: {metrics['relevant_retrieved']} chunks")
```

```python
# ================================================================
# STEP 13: Advanced - Answer Quality Evaluation
# ================================================================

def evaluate_answer_quality(question: str, answer: str, context: str):
    """Evaluate answer quality using LLM"""
    eval_prompt = f"""Evaluate this Q&A system response on a scale of 1-5 for:
1. Faithfulness (answer based on context)
2. Relevance (answer addresses question)
3. Completeness (answer is thorough)

Question: {question}

Context: {context}

Answer: {answer}

Provide scores and brief explanation in format:
Faithfulness: X/5 - explanation
Relevance: X/5 - explanation
Completeness: X/5 - explanation"""
    
    evaluation = llm.invoke(eval_prompt)
    return evaluation.content

# Example evaluation
question = "What are pooling layers?"
answer, sources = ask_question(question)
context = "\n".join([doc.page_content for doc in sources])

print("\n" + "="*60)
print("ANSWER QUALITY EVALUATION")
print("="*60)
evaluation = evaluate_answer_quality(question, answer, context)
print(evaluation)
```
